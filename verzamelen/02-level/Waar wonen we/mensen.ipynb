# Regelen met Regex

In de exercises van dit level ga je informatie extraheren uit een tekstbestand in een nieuw formaat. Het bestand heeft een lijst van namen met allerlei gegevens.

In deze exercise ga je alle bekende telefoonnummers in het tekstbestand opzoeken en bekijken. Je gaat vervolgens onderstaande vragen beantwoorden in een Jupyter Notebook:

```python
import re

numbers=[]
with open("people.txt", 'rt') as file:
#     print(file.read())
    for line in file:
        x= re.search(r"06\s\d{8}",line)
        if x != None:
            numbers.append(x.string.rstrip())

numbers
```

Ten eerste: Hoeveel telefoonnummers zijn er in het bestand te vinden?


```python
len(numbers)
```

Hoeveel telefoonnummers eindigen er met een 0?

```python
zero_ending = 0
for num in numbers:
    x = re.search(r"0$",num)
    if x:
        zero_ending += 1

zero_ending
```

Wat zijn de telefoonnummers waar het getal 666 ergens in het nummer voorkomt (dus bijv. 06 12366645 of 06 23048666 maar niet 06 66123456);


```python
for num in numbers:
    x=re.search(r"6{3}",num)
    if x:
        print(x.string)
```

Wat zijn de telefoonnummers die alleen de getallen 6, 7, 8, 9 en 0 gebruiken?


```python
for num in numbers:
    x = re.search(r".{3}[6|7|8|9|0|]{8}",num)
    if x:
        print(x.string)
```

## Waar wonen we ?

Uit dezelfde database als in de vorige exercise, ga je de andere informatie verder verwerken. Vind de volgende informatie over de postcodes die er in de database te vinden zijn:

Hoeveel mensen wonen er in Amsterdam?
(Let op! Sommige mensen wonen wel in Amsterdam, maar hebben een andere plaatsnaam in hun adres. Daarnaast zijn er ook straatnamen waar "Amsterdam" in voorkomt, dus reken die niet mee als ze niet in Amsterdam liggen!)

```python
postcodes = []
regex_key = (r"(\d){4}\s\w\w\sAmsterdam")
with open("people.txt", 'rt') as file:
    for line in file:
        if x:= re.search(regex_key,line):
            postcodes.append(x.string)

len(postcodes)
```

Wat zijn de straatnamen en huisnummers van de mensen die op postcodenummer 7311 wonen?

```python
people = ""
with open("people.txt","rt") as file:
    people = file.read()

regex_key = (r"([a-zA-Z\s]+\s[0-9]+)\s(7311\s\w\w)")
x= re.findall(regex_key,people)
for adres in x:
    print(adres[0])
```

In welke plaatsen komen straatnamen voor waarin de naam "Willem" voorkomt? Geef de plaatsen in alfabetische volgorde terug.

```python
regex_key=(r"[\w\s-]*\sWillem[\s\w-]*\s\d+\s\d{4}\s\w{2}\s(.+)")
cities = re.findall(regex_key,people)
cities = list(dict.fromkeys(cities))
print(sorted(cities))
```
